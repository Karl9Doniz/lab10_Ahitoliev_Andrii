```python
import pandas as pd
```

```python
path = '/census.csv'

def read_data(path_to_file):
    df = pd.read_csv(path_to_file)
    return df

read_data(path)
```

```python
df = read_data(path)

def max_counties(df):
    frame = df.loc[(df['SUMLEV'] == 50)]
    states = frame['STNAME'].tolist()
    occurences = {state:states.count(state) for state in states}
    return [i for i in occurences.keys() if occurences[i] == 
            max(occurences.values())][0]

max_counties(df)
```

```python
def max_difference(df):
    frame = df.loc[(df['SUMLEV'] == 50)]
    estimates = ['CTYNAME', 'POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 
            'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']
    df_pop = frame[estimates]
    population = []
    pop_dict = {}
    for index, rows in df_pop.iterrows():
        population.append(rows[estimates].tolist())
    for i in range(len(population)):
        biggest_dif = max(population[i][1:]) - min(population[i][1:])
        pop_dict[population[i][0]] = biggest_dif
    return max(pop_dict, key=pop_dict.get)

max_difference(df)
```

```python
def conditional_counties(df):
    frame = df.loc[(df['SUMLEV'] == 50)]
    df_conditional = frame.loc[((frame.REGION == 1) | (frame.REGION == 2)) 
        & (frame.POPESTIMATE2015 > frame.POPESTIMATE2014) & 
        frame['CTYNAME'].str.startswith('Washington')]
    return df_conditional[['STNAME', 'CTYNAME']]

conditional_counties(df)
```
